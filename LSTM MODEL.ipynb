```python
import math
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Bidirectional, Dropout
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
```

```python
df = pd.read_csv("STOCK_INDEX.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
close_prices = df['Close']
values = close_prices.values
training_data_len = math.ceil(len(values)* 0.8)
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(values.reshape(-1,1))
train_data = scaled_data[0:training_data_len, :]

x_train = []
y_train = []

for i in range(2, len(train_data)):
    x_train.append(train_data[i-2 : i, 0])
    y_train.append(train_data[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(64))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")
model.fit(x_train, y_train, validation_split=0.1, batch_size=32, epochs=20)

model.save('lstm_Model.h5')
```
